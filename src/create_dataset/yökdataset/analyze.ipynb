```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Load the data (abstracts of theses from 6 universities and 6 departments)
abstracts = pd.read_csv('cleaned_output/cleaned_data.csv')
```

```python
# Check the first few rows of the data
abstracts.head()
```

```python
# print total number of theses
n_theses = len(abstracts)
print(f'Total number of theses: {n_theses}')
```

```python
# create a pie chart to show the distribution of theses by university
university_counts = abstracts['university'].value_counts()
plt.pie(university_counts, labels=university_counts.index, autopct='%1.1f%%')
# add n_theses to the title 
plt.title('Distribution of Theses by University (n={})'.format(n_theses))
plt.show()
```

```python
# create a pie chart to show the distribution of theses by department (konu)
department_counts = abstracts['konu'].value_counts()
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%')
# add n_theses to the title
plt.title('Distribution of Theses by Department (n={})'.format(n_theses))
plt.show()
```

```python
# distribution of n_sentences_en (number of sentences in English abstracts)
sns.histplot(abstracts['n_sentences_en'], bins=30)
# add n_theses to the title
plt.title('Distribution of Number of Sentences in English Abstracts (n={})'.format(n_theses))
plt.show()
```
