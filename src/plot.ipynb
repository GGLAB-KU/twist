```python
import numpy as np


def smoothen_curve(line_data):
    idx = range(len(line_data))
    xnew = np.linspace(min(idx), max(idx), 100)

    spl = make_interp_spline(idx, line_data, k=5)
    smooth = spl(xnew)
    return xnew, smooth
```

```python
import pandas as pd


def compute_moving_averages(arr, window_size: int = 7):
    numbers_series = pd.Series(arr)
    windows = numbers_series.rolling(window_size)
    moving_averages = windows.mean()
    moving_averages_list = moving_averages.tolist()
    return moving_averages_list
```
