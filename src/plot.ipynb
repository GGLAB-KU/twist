```python
import numpy as np


def smoothen_curve(line_data):
    idx = range(len(line_data))
    xnew = np.linspace(min(idx), max(idx), 100)

    spl = make_interp_spline(idx, line_data, k=5)
    smooth = spl(xnew)
    return xnew, smooth
```

```python
import pandas as pd


def compute_moving_averages(arr, window_size: int = 7):
    numbers_series = pd.Series(arr)
    windows = numbers_series.rolling(window_size)
    moving_averages = windows.mean()
    moving_averages_list = moving_averages.tolist()
    return moving_averages_list
```

```python
keys = list(data.keys())
keys = list(map(lambda x: int(x.split(' - ')[0]), keys))

mt_sentence_counts = [v[0] for v in data.values()]
user_sentence_counts = [v[1] for v in data.values()]
mt_word_counts = [v[2] for v in data.values()]
user_word_counts = [v[3] for v in data.values()]
```

```python
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

use_smooth_curves = False
show_diffs = True

if use_smooth_curves:
    _, mt_sentence_counts = smoothen_curve(mt_sentence_counts)
    _, user_sentence_counts = smoothen_curve(user_sentence_counts)
    _, mt_word_counts = smoothen_curve(mt_word_counts)
    keys, user_word_counts = smoothen_curve(user_word_counts)

# fig, axs = plt.subplots(2, 1, figsize=(10, 10))

fig1, ax1 = plt.subplots(figsize=(20, 10))
fig2, ax2 = plt.subplots(figsize=(20, 10))
axs = [ax1, ax2]

if show_diffs:
    sentence_diff = np.array(user_sentence_counts) - np.array(mt_sentence_counts)
    sentence_diff = sentence_diff / np.array(user_sentence_counts)
    word_diff = np.array(user_word_counts) - np.array(mt_word_counts)
    word_diff = word_diff / np.array(user_word_counts)
    sentence_diff = compute_moving_averages(sentence_diff, window_size=10)
    word_diff = compute_moving_averages(word_diff, window_size=10)
    # axs[0].set_yscale('log')
    # axs[0].set_xscale('log')
    axs[0].plot(keys, sentence_diff, label='Normalized Sentence Count Difference (User - MT)', marker='o')
    axs[0].set_title('Sentence Count Differences')
    axs[0].set_xlabel('Word Count of User Text')
    axs[0].set_ylabel('Sentence Diff. / User Sentence Count')
    axs[0].legend()
    fig1.savefig('../data/sentence_diff.png')
    
    # axs[1].set_xscale('log')
    axs[1].plot(keys, word_diff, label='Normalized Word Count Difference (User - MT)', marker='o')
    axs[1].set_title('Word Count Differences')
    axs[1].set_xlabel('Word Count of User Text')
    axs[1].set_ylabel('Word Diff. / User Word Count')
    axs[1].legend()
    fig2.savefig('../data/word_diff.png')
    
else:
    axs[0].set_yscale('log')
    axs[0].set_xscale('log')
    axs[0].plot(keys, mt_sentence_counts, label='MT Sentence Count', marker='o')
    axs[0].plot(keys, user_sentence_counts, label='User Sentence Count', marker='s')
    axs[0].set_title('Sentence Count Comparison')
    axs[0].set_xlabel('Text Length Range')
    axs[0].set_ylabel('Sentence Count')
    axs[0].legend()

    axs[1].set_yscale('log')
    axs[1].set_xscale('log')
    axs[1].plot(keys, mt_word_counts, label='MT Word Count', marker='o')
    axs[1].plot(keys, user_word_counts, label='User Word Count', marker='s')
    axs[1].set_title('Word Count Comparison')
    axs[1].set_xlabel('Text Length Range')
    axs[1].set_ylabel('Word Count')
    axs[1].legend()

plt.tight_layout()
plt.show()
```
