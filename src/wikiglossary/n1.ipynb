```python
# write a code for csv to excel
import pandas as pd
df = pd.read_csv('all_glossary_with_wikipedia.csv')
df.to_excel('n1.xlsx', index=False)
```

```python
import matplotlib.pyplot as plt

# Data to visualize
total_items = 57
selected_items = 35

# Create a pie chart
labels = 'Terimler.org', 'Not in Terimler.org'
sizes = [selected_items, total_items - selected_items]
colors = ['#ff9999','#66b3ff']
explode = (0.1, 0)  # explode the 1st slice

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Percentage of Terms in Terimler.org')
plt.show()
```
